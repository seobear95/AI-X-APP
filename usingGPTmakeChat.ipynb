2.  GPT-2 모델 및 토크나이저 로딩
ransformers 라이브러리에서 GPT2 모델과 토크나이저를 로드합니다. 여기에서는 'gpt2'라는 모델을 사용

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

3. 입력 문장 토크나이징
용자의 입력을 받아 토크나이즈하는 과정
용자의 문장을 모델이 이해할 수 있는 형태로 변환하는 과정

```python
inputs = tokenizer.encode("what is chatGPT?", return_tensors='pt')
```

4. GPT-2를 이용한 응답 생성
토크나이징된 입력을 바탕으로 모델을 사용
응답을 생성
최대 150 글자의 3개 다른 응답을 생성하도록 설정

```python
outputs = model.generate(inputs, max_length=150, num_return_sequences=1,
                         no_repeat_ngram_size=2, temperature=0.7)
```

5. 생성된 응답 디코딩
마지막으로, 생성된 응답을 디코딩
이해할 수 있는 문장으로 변환
챗봇이 사용자의 입력에 어떤 응답을 생성하는지 확인

```python
for i in range(1):
    print(f"Generated response {i + 1}: {tokenizer.decode(outputs[i], skip_special_tokens=True)}")
```
